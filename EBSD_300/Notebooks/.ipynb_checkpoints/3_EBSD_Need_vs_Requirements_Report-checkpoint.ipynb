# System Engineering Generative AI Enabled Knowledge Partner

This notebook demonstrates how a System Engineering Generative AI Knowledge Partner can assist with the analysis of stakeholder needs against an existing requirements baseline. It is designed to help systems engineers reason across requirement data using AI in a structured and explainable way, without replacing engineering judgment.

In this workflow, the notebook compares a table of identified needs with an established set of system requirements and looks for requirements that may be directly impacted, indirectly affected, or that may constrain delivery of the need. The emphasis is not limited to simple text matching. Instead, the Knowledge Partner applies systems engineering reasoning to identify broader relationships, such as how a performance-related need may influence processor activity, sensing rates, communication activity, power demand, thermal behavior, duty cycle, or runtime constraints.

The notebook is especially useful in early assessment and change analysis activities, where engineers want to understand how a new or evolving need could ripple across an existing requirement set. It helps expose both obvious and less visible dependencies, including second-order effects such as performance changes leading to increased power consumption or tighter thermal limits.

The output is structured in a clear tabular form so that engineers can review the identified requirement impacts, understand the rationale behind each linkage, and use the results as a starting point for refinement, review, or downstream architecture analysis. This makes the notebook valuable not only as an analysis aid, but also as a practical example of how AI can be used as a reasoning partner within an MBSE-oriented workflow.

At a broader level, this notebook illustrates an important role for the Knowledge Partner: helping transform requirement and needs data into actionable engineering insight. Rather than simply retrieving content, it supports trace-oriented thinking, impact awareness, and informed decision-making in a way that complements the engineer’s expertise.
