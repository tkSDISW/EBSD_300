# SE Knowledge Partner  Requirements Impact to Physcal Components

This notebook demonstrates how the System Engineering Knowledge Partner can analyze the effect of requirement changes on architecture elements represented in a system model fabric. Its focus is on requirements that are linked to PhysicalComponent Node components and on using the surrounding model context to explain why those components would be impacted by change.

The workflow begins by locating requirements in the fabric that are connected to PhysicalComponent Node elements through model relationships such as satisfaction or other requirement traces present in the fabric. Once those requirement-to-component links are identified, the notebook uses the same fabric content to build a detailed engineering rationale for impact. Rather than stopping at traceability, it examines the architectural role of the component, along with related ports, interfaces, exchanges, functions, constraints, and neighboring elements, to explain how and why a requirement change could affect the design.

This makes the notebook especially useful for change-impact assessment. It helps expose not only that a physical component is tied to a requirement, but also what aspect of the architecture may need reconsideration, such as physical layout, thermal behavior, power delivery, timing, capacity, or interface design. The notebook therefore turns requirement trace data into more meaningful system insight by combining model relationships with engineering reasoning.

At a broader level, this notebook illustrates the value of the Knowledge Partner as more than a retrieval tool. It acts as a reasoning partner over model-derived fabric, helping engineers move from linked data to justified impact understanding. That makes it a practical example of how AI can support architecture-aware change analysis within an MBSE workflow.
